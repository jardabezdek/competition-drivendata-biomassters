```python
import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf
import tifffile as tiff
```

```python
CHIP_ID = "00a28320"
SPLIT = "test"
MONTH_INDEX = "00"
```

## Features

### S1

```python
# read image

img = tiff.imread(f"/usr/src/app/data/{SPLIT}_features/{CHIP_ID}_S1_{MONTH_INDEX}.tif")
img.shape
```

```python
# show all the 4 bands for S1 image

for i in range(4):

    plt.imshow(
        tf.transpose(img, perm=(2, 0, 1)).numpy()[i], 
        interpolation="nearest"
    )
    plt.show()
```

### S2

```python
# read image

img = tiff.imread(f"/usr/src/app/data/{SPLIT}_features/{CHIP_ID}_S2_{MONTH_INDEX}.tif")
img.shape
```

```python
# show all the 11 bands for S2 image

for i in range(11):

    plt.imshow(
        tf.transpose(img, perm=(2, 0, 1)).numpy()[i], 
        interpolation="nearest"
    )
    plt.show()
```
